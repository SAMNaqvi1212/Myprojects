```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
scaling=ss.fit_transform(data)
```

```python
ss_data=pd.DataFrame(scaling,columns=data.columns)
```

```python
ss_data.head()
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(ss_data)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters =3 , init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(ss_data)
```

```python
pd.Series(kmeans.labels_).value_counts()
```

```python
preds=kmeans.labels_
df["Kmeans_clusters"]=preds
```

```python
df.head()
```

```python
sns.scatterplot("child_mort","gdpp",data=df,hue="Kmeans_clusters")
```

```python
sns.scatterplot("inflation","gdpp",data=df,hue="Kmeans_clusters")
```

```python
developed_countries=df[df["Kmeans_clusters"]==0]["country"]
developing_countries=df[df["Kmeans_clusters"]==1]["country"]
under_developed_countries=df[df["Kmeans_clusters"]==2]["country"]
```

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(ss_data, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters =3 , affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(ss_data)
```

```python
preds_1=hc.labels_
```
